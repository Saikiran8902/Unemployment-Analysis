```python
def descriptive_stats(country_code):
    # Filter the dataframe to only include the specified country
    country_df = df[df['Country_code'] == country_code]

    # Calculate mean, median, mode, and standard deviation of the unemployment rate
    mean = country_df['No_of_Unemploymnet'].mean()
    median = country_df['No_of_Unemploymnet'].median()
    mode = country_df['No_of_Unemploymnet'].mode()
    std_dev = country_df['No_of_Unemploymnet'].std()

    # Print the results
    print(f"Descriptive statistics for {country_code}:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Mode: {mode[0]:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")

# Call the function with the desired country code
a=input()
descriptive_stats(a)
```

```python
def min_max_year(country_code):
    # Filter the dataframe to only include the specified country
    country_df = df[df['Country_code'] == country_code]

    # Find the year with minimum and maximum unemployment rate
    min_year = country_df.loc[country_df['No_of_Unemploymnet'].idxmin()]['Year']
    max_year = country_df.loc[country_df['No_of_Unemploymnet'].idxmax()]['Year']

    # Print the results
    print(f"For {country_code}:")
    print(f"Year with minimum unemployment rate: {min_year}")
    print(f"Year with maximum unemployment rate: {max_year}")

# Call the function with the desired country code
b=input()
min_max_year(b)
```

```python
countries = df['Country_code'].unique()

# Create a new dataframe to store the predicted values
predictions = pd.DataFrame(columns=['Country', 'Year', 'Value'])
for country in countries:
    # Get the data for the current country
    country_data = df[df['Country_code'] == country].reset_index(drop=True)
    
    # Calculate the moving average with a 3-month window
    ma_values = country_data['No_of_Unemploymnet'].rolling(window=3).mean().iloc[-1:]
    
    # Predict the next 2 years' values using the MA model
    ma_predictions = pd.DataFrame({'Value': np.repeat(ma_values, 8)})
    ma_predictions.index = pd.date_range(start='2022-01-01', end='2023-12-01', freq='QS')
    
    # Add the predictions to the dataframe
    for year in range(2022, 2024):
        predictions = predictions.append({'Country': country, 'Year': year, 'Value': ma_predictions.loc[str(year)].values[0]}, ignore_index=True)

# Calculate the Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs(predictions['Value'].values - np.mean(predictions['Value'].values)) / np.mean(predictions['Value'].values)) * 100
l=list(map('{:.2f}%'.format,mape))
l
```

```python
from statsmodels.tsa.arima.model import ARIMA
def predict_arima(country):
    # Subset data for the given country
    country_data = df[df['Country_name']==country]
    # Convert the date column to datetime format
    # Set date as the index of the dataframe
    # Fit ARIMA model on the unemployment rate column
    model = ARIMA(country_data['No_of_Unemploymnet'], order=(1,1,1))
    results = model.fit()
    # Make predictions for the next 2 years
    forecast = results.forecast(steps=8)
    # Convert the forecasted values to a dataframe
    forecast_df = pd.DataFrame({'Date': pd.date_range(start='2022-01-01', periods=8, freq='YS'),
                                'Country': country,
                                'Unemployment Rate': forecast})
    return forecast_df

# List of countries in the dataset
countries = df['Country_name'].unique()

# Initialize an empty dataframe to store the predicted values
predictions = pd.DataFrame(columns=['Date', 'Country', 'Unemployment Rate'])

# Loop through all the countries and make predictions
for country in countries:
    country_predictions = predict_arima(country)
    predictions = pd.concat([predictions, country_predictions])

# Print the predicted values
print(predictions)
```

```python
max_values = {}
min_values = {}

# Loop through all the countries and make predictions
for country in countries:
    # Make predictions for the current country
    predictions = predict_arima(country)
    # Loop through each of the next 2 years (2022 and 2023)
    for year in range(2022, 2024):
        # Subset the predicted values for the current year and country
        year_predictions = predictions[predictions['Date'].dt.year == year]
        # Find the maximum and minimum values for the current year and country
        max_value = year_predictions['Unemployment Rate'].max()
        min_value = year_predictions['Unemployment Rate'].min()
        # Store the maximum and minimum values in the dictionaries
        if country not in max_values:
            max_values[country] = {}
        if country not in min_values:
            min_values[country] = {}
        max_values[country][year] = max_value
        min_values[country][year] = min_value

# Print the maximum and minimum values for each country in each of the next 2 years
print('Maximum Unemployment Rate')
print(pd.DataFrame(max_values).T)
print('\nMinimum Unemployment Rate')
print(pd.DataFrame(min_values).T)
```
